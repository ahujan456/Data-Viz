```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
```

```python
df = pd.read_csv('task_data.csv')
```

```python
df.head()
```

```python
# df.groupby('Country')['PM2.5'].sum()
```

```python
iran_plot = df[df['Country']=='Iran'].groupby('Year')['PM2.5'].sum()
china_plot = df[df['Country']=='China'].groupby('Year')['PM2.5'].sum()
```

```python
iran_plot
plt.plot(iran_plot, label='Iran')
plt.plot(china_plot, label='China')

plt.title("Yearwise China and Iran PM2.5 Comparison")
plt.xticks(df['Year'].value_counts().index)
plt.ylabel('PM2.5')
plt.xlabel("Year")
plt.legend()
```

```python
plt.hist(df['PM10'], bins=50, facecolor='green', alpha=0.6)
plt.title('PM10 Histogram')
plt.ylabel("Probablity")
plt.xlabel("Bins")
plt.grid()
```

```python
cols = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for col in cols:
    plt.plot(sdf['month_number'], sdf[col], label=col)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(sdf['month_number'])
plt.title('Sales data')
plt.grid()
```

```python
sdf['date'] = pd.to_datetime(['2020-{}-01'.format(month) for month in sdf['month_number']])
```

```python
sdf['quarter'] = sdf['date'].dt.quarter
sdf.drop('date', axis=1, inplace=True)
```

```python
final_sdf = sdf.groupby('quarter')
final_df = final_sdf.sum()
```

```python
i = -1
for col in final_df.columns[1:7]:
  plt.bar(final_df.index + i,final_df[col],width=0.15,label=col)
  i = i - 0.15

plt.xticks(final_df.index-1.4,final_df.index)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
